```python
import pandas as pd
from pathlib import Path

poverty_2010 = pd.read_csv("Education_level_yearly_data/Ed_2010.csv")
poverty_2011 = pd.read_csv("Education_level_yearly_data/Ed_2011.csv")
poverty_2012 = pd.read_csv("Education_level_yearly_data/Ed_2012.csv")
poverty_2013 = pd.read_csv("Education_level_yearly_data/Ed_2013.csv")
poverty_2014 = pd.read_csv("Education_level_yearly_data/Ed_2014.csv")

poverty_2010
```

```python
df = [
    (poverty_2010, "2010"),
    (poverty_2011, "2011"),
    (poverty_2012, "2012"),
    (poverty_2013, "2013"),
    (poverty_2014, "2014")]
```

```python
def clean(df):
    clean_df = []
    for dataframe, year in df:
            dataframe = dataframe.drop(columns=[column for column in dataframe.columns if column.endswith('M')])
            dataframe = dataframe.drop(columns=["GEO_ID"])
            dataframe = dataframe.drop(columns=dataframe.columns[dataframe.apply(lambda col: col.astype(str).str.contains('\(X\)').any())])
            dataframe["NAME"] = dataframe["NAME"].str.replace(', California|city|CDP|town', '', regex=True)
            dataframe["NAME"] = dataframe["NAME"].str.strip()
            dataframe["Year"] = year   
            dataframe = dataframe[["Year", "NAME", "DP02_0059E", "DP02_0059PE", "DP02_0060E",
                                   "DP02_0060PE", "DP02_0061E", "DP02_0061PE", "DP02_0062E",
                                   "DP02_0062PE", "DP02_0063E", "DP02_0063PE", "DP02_0064E",
                                   "DP02_0064PE", "DP02_0065E", "DP02_0065PE"]]
            dataframe = dataframe.rename(columns={
                "NAME": "City",
                "DP02_0059E": "Population 25 years and over - Less than 9th grade",
                "DP02_0059PE": "Percent Population 25 years and over - Less than 9th grade",
                "DP02_0060E": "Population 25 years and over - 9th to 12th (No Diploma)",
                "DP02_0060PE": "Percent Population 25 years and over - 9th to 12th (No Diploma)",
                "DP02_0061E": "Population 25 years and over - High School Graduate (and equivalent)",
                "DP02_0061PE": "Percent Population 25 years and over - High School Graduate (and equivalent)",
                "DP02_0062E": "Population 25 years and over - Some college, no degree",
                "DP02_0062PE": "Percent Population 25 years and over - Some college, no degree",
                "DP02_0063E": "Population 25 years and over - Associate's degree",
                "DP02_0063PE": "Percent Population 25 years and over - Associate's degree",
                "DP02_0064E": "Population 25 years and over - Bachelor's degree",
                "DP02_0064PE": "Percent Population 25 years and over - Bachelor's degree",
                "DP02_0065E": "Population 25 years and over - Graduate or Prefessional Degree",
                "DP02_0065PE": "Percent Population 25 years and over - Graduate or Prefessional Degree"})
            dataframe = dataframe.drop(dataframe.index[0])
            clean_df.append(dataframe)
    return clean_df
cleaned_df = clean(df)
```

```python
years = ["2010", "2011", "2012", "2013", "2014"]
file_name = {f"Cleaned_Ed_{year}": df for df, year in zip(cleaned_df, years)}
for name, df in file_name.items():
    csv_path= f"{name}.csv"
    df.to_csv(csv_path, index=False)
```
