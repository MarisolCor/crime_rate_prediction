```python
import pandas as pd
import re
from pathlib import Path
data=pd.read_csv(Path("Necessary Education Level/Education_Data.csv"))

data.head()
```

```python
def cleanCityNames(df):
    df['Geographic Area Name'] = df['Geographic Area Name'].str.replace(', California|city|town|CDP', '', regex=True)
    df['Geographic Area Name']=  df['Geographic Area Name'].str.strip()
```

```python
cleanCityNames(data)
data.head()
```

```python
data.to_csv('clean_education_level.csv',index=False)
```
