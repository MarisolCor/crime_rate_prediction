```python
import pandas as pd
import re
from pathlib import Path

```

```python
def cleanCityNames(df):
    df['City'] = df['City'].str.replace(', California|city|town|CDP', '', regex=True)
    df['City']=  df['City'].str.strip()
    
def addYear(df,year):
    df.insert(0,'Year',year)
    
def cleanN(df):
    cleanCityNames(df)
    condition = df['Unemployment Rate'] == 'N'
    df.drop(df[condition].index, inplace=True)
    condition2 = df['City'] == 'California'
    df.drop(df[condition2].index, inplace=True)
```

```python
columns_keep=['NAME','DP03_0009PE','DP03_0051E']
new_column_names={'NAME':'City','DP03_0009PE':'Unemployment Rate','DP03_0051E':'Income'}
```

```python
df_data_2010=pd.read_csv('resources/ACSDP1Y2010.DP03-Data.csv')
df_data_2011=pd.read_csv('resources/ACSDP1Y2011.DP03-Data.csv')
df_data_2012=pd.read_csv('resources/ACSDP1Y2012.DP03-Data.csv')
df_data_2013=pd.read_csv('resources/ACSDP1Y2013.DP03-Data.csv')
df_data_2014=pd.read_csv('resources/ACSDP1Y2014.DP03-Data.csv')
df_data_2015=pd.read_csv('resources/ACSDP1Y2015.DP03-Data.csv')
df_data_2016=pd.read_csv('resources/ACSDP1Y2016.DP03-Data.csv')
df_data_2017=pd.read_csv('resources/ACSDP1Y2017.DP03-Data.csv')
df_data_2018=pd.read_csv('resources/ACSDP1Y2018.DP03-Data.csv')
df_data_2019=pd.read_csv('resources/ACSDP1Y2019.DP03-Data.csv')

df_data_2010.head()
```

```python
df_data_2010=df_data_2010[columns_keep]
df_data_2011=df_data_2011[columns_keep]
df_data_2012=df_data_2012[columns_keep]
df_data_2013=df_data_2013[columns_keep]
df_data_2014=df_data_2014[columns_keep]
df_data_2015=df_data_2015[columns_keep]
df_data_2016=df_data_2016[columns_keep]
df_data_2017=df_data_2017[columns_keep]
df_data_2018=df_data_2018[columns_keep]
df_data_2019=df_data_2019[columns_keep]
df_data_2010
```

```python
df_data_2010=df_data_2010.rename(columns=new_column_names).drop(0)
df_data_2011=df_data_2011.rename(columns=new_column_names).drop(0)
df_data_2012=df_data_2012.rename(columns=new_column_names).drop(0)
df_data_2013=df_data_2013.rename(columns=new_column_names).drop(0)
df_data_2014=df_data_2014.rename(columns=new_column_names).drop(0)
df_data_2015=df_data_2015.rename(columns=new_column_names).drop(0)
df_data_2016=df_data_2016.rename(columns=new_column_names).drop(0)
df_data_2017=df_data_2017.rename(columns=new_column_names).drop(0)
df_data_2018=df_data_2018.rename(columns=new_column_names).drop(0)
df_data_2019=df_data_2019.rename(columns=new_column_names).drop(0)
df_data_2015
```

```python

cleanN(df_data_2010)
cleanN(df_data_2011)
cleanN(df_data_2012)
cleanN(df_data_2013)
cleanN(df_data_2014)
cleanN(df_data_2015)
cleanN(df_data_2016)
cleanN(df_data_2017)
cleanN(df_data_2018)
cleanN(df_data_2019)

df_data_2015
```

```python
addYear(df_data_2010, '2010')
addYear(df_data_2011, '2011')
addYear(df_data_2012, '2012')
addYear(df_data_2013, '2013')
addYear(df_data_2014, '2014')
addYear(df_data_2015, '2015')
addYear(df_data_2016, '2016')
addYear(df_data_2017, '2017')
addYear(df_data_2018, '2018')
addYear(df_data_2019, '2019')
```
