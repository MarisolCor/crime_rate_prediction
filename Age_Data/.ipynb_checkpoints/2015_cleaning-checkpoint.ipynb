```python
import pandas as pd
```

```python
# read file
file_path = 'Age_2015.csv'

df = pd.read_csv(file_path)

print(df)
```

```python
# remoove the unwanted rows
df = df[~df['Label (Grouping)'].str.contains('Margin of Error|Percent Margin of Error', case=False)]

print(df)
```

```python
# save the changes to the csv file
df.to_csv('Age_2015.csv', index=False)
print('changes saved')
```
