```python
import pandas as pd
from pathlib import Path

age_2010 = pd.read_csv("Age_Yearly/Age_2010.csv")
age_2011 = pd.read_csv("Age_Yearly/Age_2011.csv")
age_2012 = pd.read_csv("Age_Yearly/Age_2012.csv")
age_2013 = pd.read_csv("Age_Yearly/Age_2013.csv")
age_2014 = pd.read_csv("Age_Yearly/Age_2014.csv")

age_2010

```

```python
df = [
    (age_2010, "2010"),
    (age_2011, "2011"),
    (age_2012, "2012"),
    (age_2013, "2013"),
    (age_2014, "2014")]
```

```python
def clean(df):
    clean_df = []
    for dataframe, year in df:
            dataframe = dataframe.drop(columns=["GEO_ID"])
            dataframe["NAME"] = dataframe["NAME"].str.replace(', California|city|CDP|town', '', regex=True)
            dataframe["NAME"] = dataframe["NAME"].str.strip()
            dataframe["Year"] = year   
            dataframe = dataframe[["Year", "NAME", "DP05_0001E", "DP05_0004E", "DP05_0004PE",
                                   "DP05_0005E", "DP05_0005PE", "DP05_0006E", "DP05_0006PE",
                                   "DP05_0007E", "DP05_0007PE", "DP05_0008E", "DP05_0008PE",
                                   "DP05_0009E", "DP05_0009PE", "DP05_0010E", "DP05_0010PE",
                                   "DP05_0011E", "DP05_0011PE", "DP05_0012E", "DP05_0012PE",
                                   "DP05_0013E", "DP05_0013PE", "DP05_0014E", "DP05_0014PE", 
                                   "DP05_0015E", "DP05_0015PE", "DP05_0016E", "DP05_0016PE"]]
            dataframe = dataframe.rename(columns={
                "NAME": "City",
                "DP05_0001E": "Total population",
                "DP05_0004E": "Under 5 years",
                "DP05_0004PE": "Percent Under 5 years",
                "DP05_0005E": "5 to 9 years",
                "DP05_0005PE": "Percent 5 to 9 years",
                "DP05_0006E": "10 to 14 years",
                "DP05_0006PE": "Percent 10 to 14 years",
                "DP05_0007E": "15 to 19 years",
                "DP05_0007PE": "Percent 15 to 19 years",
                "DP05_0008E": "20 to 24 years",
                "DP05_0008PE": "Percent 20 to 24 years",
                "DP05_0009E": "25 to 34 years",
                "DP05_0009PE": "Percent 25 to 34 years",
                "DP05_0010E": "35 to 44 years",
                "DP05_0010PE": "Percent 35 to 44 years",
                "DP05_0011E": "45 to 54 years",
                "DP05_0011PE": "Percent 45 to 54 years",
                "DP05_0012E": "55 to 59 years",
                "DP05_0012PE": "Percent 55 to 59 years",
                "DP05_0013E": "60 to 64 years",
                "DP05_0013PE": "Percent 60 to 64 years",
                "DP05_0014E": "65 to 74 years",
                "DP05_0014PE": "Percent 65 to 74 years",
                "DP05_0015E": "75 to 84 years",
                "DP05_0015PE": "Percent 75 to 84 years",
                "DP05_0016E": "85 years and over",
                "DP05_0016PE": "Percent 85 years and over",
            })
            dataframe = dataframe.drop(dataframe.index[0])
            clean_df.append(dataframe)
    return clean_df
cleaned_df = clean(df)
```

```python
years = ["2010", "2011", "2012", "2013", "2014"]
file_name = {f"Cleaned_Age_{year}": df for df, year in zip(cleaned_df, years)}
for name, df in file_name.items():
    csv_path= f"{name}.csv"
    df.to_csv(csv_path, index=False)
```
