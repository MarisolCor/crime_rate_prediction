```python
import pandas as pd
import re
from pathlib import Path
```

```python
edu_data=pd.read_csv(Path("education_level_california/clean_education_level.csv"))
edu_data = edu_data.rename(columns={'Geographic Area Name': 'City'})
edu_data.head()
```

```python
age_data=pd.read_csv(Path("Age_Data/Merged_Age.csv"))

age_data.head()
```

```python
crime_data=pd.read_csv(Path("crime_data/crime_cleaned.csv"))

crime_data.head()
```

```python
income_data=pd.read_csv(Path("data_incomes_california_cities/clean_data/incomes_2015to2019.csv"))

income_data.head()
```

```python
income_data=pd.read_csv(Path("Poverty_Resources/All_Family_Poverty.csv"))

income_data.head()
```

```python
house_price_data=pd.read_csv(Path("Poverty_Resources/Clean_House_Price.csv"))
house_price_data = house_price_data.rename(columns={'RegionName': 'City'})
house_price_data.head()
```

```python
rent_own_data=pd.read_csv(Path("rent_vs_own_data/rent_vs_own_cleaned.csv"))

rent_own_data.head()
```

```python
# edu_data age_data crime_data income_data house_price_data rent_own_data
all_df = edu_data.merge(age_data, on=['City', 'Year'], how='left') \
                 .merge(crime_data, on=['City', 'Year'], how='left') \
                 .merge(income_data, on=['City', 'Year'], how='left') \
                 .merge(house_price_data, on=['City', 'Year'], how='left') \
                 .merge(rent_own_data, on=['City', 'Year'], how='left')
```

```python
all_df
```

```python
all_df_clean = all_df.dropna()
```

```python
all_df_reset = all_df_clean.reset_index(drop=True)
```

```python
all_df_reset
```

```python
all_df_reset.to_csv('All_data_cleaned.csv', index=False)
```
